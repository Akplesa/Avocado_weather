```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

file_to_load = "Output/weather.csv"

caweather_df = pd.read_csv(file_to_load)
caweather_df.head()
```

```python
caweather_df = caweather_df[["Severity","EndTime(UTC)","County","State"]]
caweather_df
```

```python
caweather_df.columns=[["Severity","Date","County", "State"]]
caweather_df.head()
```

```python
caweather_df = caweather_df.rename(columns={'Severity': 'severity', 'Date': 'date','County':'county'})
caweather_df
```
