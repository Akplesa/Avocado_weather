```python
 import pandas as pd
from sqlalchemy import create_engine
```

```python
drought_file = "Resources/Drought.csv"
drought_df = pd.read_csv(drought_file)
drought_df.head()
```

```python
#selecting only the columns with useful info
drought_cols = ["releaseDate", "county", "state", "D3", "D4"]
drought_transformed= drought_df[drought_cols].copy()

# Rename the column headers
CA_drought = drought_transformed.rename(columns={"releaseDate": "date",
                                                          "county": "county",
                                                          "state": "state", 
                                                         "D3": "drought_level_extreme",
                                                         "D4":"drought_level_exceptional"})

CA_drought.head()
```

```python
#clean up by county, and interested in Los Angeles, San Francisco, San Diego, Sacramento
CA_drought_by_county = CA_drought.loc[(CA_drought["county"] == "Los Angeles County") | (CA_drought["county"] == "San Francisco County") | (CA_drought["county"] == "Sacramento County") | (CA_drought["county"] == "San Diego County")]
#CA_drought_by_county = CA_drought_by_county.reset_index(drop=False)
#del CA_drought_by_county['index']
CA_drought_by_county.head(50)









```

```python
connection_string = "postgres:924runner@localhost:5432/Avocado_weather"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# # Confirm tables
engine.table_names()

```

```python
CA_drought_by_county.to_sql(name='drought', con=engine, if_exists='append', index=True) 
```
