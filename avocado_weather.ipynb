```python
 import pandas as pd
from sqlalchemy import create_engine
```

```python
drought_file = "Resources/Drought.csv"
drought_df = pd.read_csv(drought_file)
drought_df.head()
```

```python
#selecting only the columns with useful info
drought_cols = ["releaseDate", "county", "state", "D3", "D4"]
drought_transformed= drought_df[drought_cols].copy()

# Rename the column headers
drought_transformed = drought_transformed.rename(columns={"releaseDate": "Date",
                                                          "county": "County",
                                                          "state": "State", 
                                                         "D3": "Extreme",
                                                         "D4":"Exceptional"})

drought_transformed.head()
```

```python
#clean up by state, and interested in CA
CA_drought = drought_transformed[drought_transformed["State"] == "CA"]
CA_drought
```
